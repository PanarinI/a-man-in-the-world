```python
import pygame
import random


# Инициализация Pygame
pygame.init()

# Параметры экрана
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Масштаб: от человека к 7 миллиардам")
clock = pygame.time.Clock()

# Цвета
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
ORANGE = (255, 165, 0)
GRAY = (150, 150, 150)

# Узлы и параметры уровней
levels = [
    {"name": "Я", "count": 0, "color": ORANGE, "size": 10},  # Уровень 1: только "Я"
    {"name": "Семья", "count": 10, "color": GRAY, "size": 8},  # Уровень 2: 10 точек
    {"name": "Сообщество", "count": 50, "color": WHITE, "size": 6},  # Уровень 3: 50 точек
    {"name": "Город", "count": 200, "color": WHITE, "size": 4},  # Уровень 4: 200 точек
    {"name": "Страна", "count": 1000, "color": WHITE, "size": 2},  # Уровень 5: 1000 точек
    {"name": "Мир", "count": 5000, "color": WHITE, "size": 1},  # Уровень 6: 5000 точек
]
current_level = 0  # Текущий уровень (начинаем с 0)

# Центральная точка
center_node = {"pos": (WIDTH // 2, HEIGHT // 2), "size": 10, "color": ORANGE}

# Функция для генерации точек


def generate_nodes(level_data):
    """Генерирует узлы для текущего уровня."""
    nodes = [center_node]  # Всегда добавляем центральный узел
    count = level_data["count"]  # Количество дополнительных точек
    size = level_data["size"]  # Размер точек
    color = level_data["color"]  # Цвет точек

    if count <= 100:  # Если узлов немного, равномерное распределение по кругу
        center_x, center_y = center_node["pos"]
        radius = 100 + current_level * 50
        for i in range(count):
            angle = i * (360 / count)
            x = center_x + radius * random.uniform(0.9, 1.1) * random.uniform(-1, 1)
            y = center_y + radius * random.uniform(0.9, 1.1) * random.uniform(-1, 1)
            nodes.append({"pos": (x, y), "size": size, "color": color})
    else:  # Для крупных уровней равномерно распределяем точки по всему экрану
        for _ in range(count):
            x = random.randint(0, WIDTH)
            y = random.randint(0, HEIGHT)
            nodes.append({"pos": (x, y), "size": size, "color": color})
    return nodes


# Изначальные узлы
nodes = generate_nodes(levels[current_level])


def draw_graph():
    """Рисует текущий граф."""
    for node in nodes:
        pygame.draw.circle(screen, node["color"], (int(node["pos"][0]), int(node["pos"][1])), node["size"])


def change_level(delta):
    """Изменяет уровень масштаба."""
    global current_level, nodes
    new_level = current_level + delta
    if 0 <= new_level < len(levels):  # Проверяем, чтобы уровень был в пределах
        current_level = new_level
        nodes = generate_nodes(levels[current_level])


# Основной цикл
running = True
while running:
    screen.fill(BLACK)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEWHEEL:
            if event.y > 0:  # Скролл вперед
                change_level(1)
            elif event.y < 0:  # Скролл назад
                change_level(-1)

    draw_graph()
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```
